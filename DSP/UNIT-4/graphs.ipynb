```python
import numpy as np
 # First create some toy data:
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.sin(x**2)
y3 = y1**2
y4 = y2**2
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y1,color="red",marker="*")
ax[0,1].plot(x,y2,color="green",linestyle='dotted')
ax[1,0].plot(x,y3,color="violet",linestyle='dashed',linewidth=5)
ax[1,1].plot(x,y4,color="black",linestyle='-.')
#linestyle=can have -.,-,--,:,'solid','dashed','None','',' ','dotted'
ax[0, 0].set_title('Simple plot with sin(x)')
ax[0, 1].set_title('Simple plot with sin(x**2)')
ax[1, 0].set_title('Simple plot with sin(x)**2')
ax[1, 1].set_title('Simple plot with sin(x**2)**2')
fig.suptitle('Stacked subplots in two direction')
plt.show()
```

```python
# Group 1 data
group1_x = [1, 2]
group1_y = [10, 11]
group1_text = ['<b>Group 1 - A</b><br>Details: High priority', '<b>Group 1 - B</b><br>Details: Medium priority']

# Group 2 data
group2_x = [3, 4]
group2_y = [12, 13]
group2_text = ['<b>Group 2 - C</b><br>Details: Low priority', '<b>Group 2 - D</b><br>Details: Critical']

fig = go.Figure()

# Add Group 1
fig.add_trace(go.Scatter(
    x=group1_x, y=group1_y,
    mode='markers',
    name='Group 1',
    text=group1_text,
    hoverinfo='text+name',
    marker=dict(
        color='blue',
        size=[40, 60],
        opacity=[0.9, 0.8]
    )
))

# Add Group 2
fig.add_trace(go.Scatter(
    x=group2_x, y=group2_y,
    mode='markers',
    name='Group 2',
    text=group2_text,
    hoverinfo='text+name',
    marker=dict(
        color='red',
        size=[80, 30],
        opacity=[0.6, 1]
    )
))

# Layout customization
fig.update_layout(
    title='Interactive Scatter Plot with Multiple Groups',
    xaxis_title='X Value',
    yaxis_title='Y Value',
    legend_title='Groups',
    template='plotly_white'
)

# Save as HTML aand PNG
fig.write_html('scatter_plot.html')
fig.write_image("scatter_plot.png")
fig.show()
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
iris = datasets.load_iris()
X = iris.data  
y = iris.target  
# Feature selection
x = X[:, 0]  # sepal length
y_ = X[:, 1]  # sepal width
z = X[:, 2]  # petal length
fig = plt.figure() #a blank sheet for drawing
ax = fig.add_subplot(111, projection='3d')


colors = ['orange', 'magenta', 'blue']
labels = iris.target_names
#Gets the species names: ['setosa', 'versicolor', 'virginica']
for i in range(3):
    ax.scatter(
        x[y == i], y_[y == i], z[y == i],
        color=colors[i], label=labels[i], s=100, edgecolors='k' #black egde
    )
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
ax.set_title("3D Iris Dataset Visualization")
ax.legend()
plt.show()
```
